Train KMeans Clustering

```python
data1=data.copy(deep=True)
```

```python
clust=data1.drop(['Gender','Cust_ID','Orders'],axis=1)
clust.head()
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaled = StandardScaler().fit(clust)       
features = scaled.transform(clust)
```

```python
features_scaled = pd.DataFrame( features, columns= clust.columns[0:36])
features_scaled.head()
```

We dont know which cluster is correct.So we create clusters from the range of 3 to 14 then find the correct cluster using silhouette_score

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
for i in range(3,15):
    kmeans = KMeans(n_clusters=i, max_iter=50)
    kmeans.fit(features_scaled)
    cluster_labels = kmeans.labels_
    silhouette_avg = silhouette_score(features_scaled, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(i, silhouette_avg))
```

From the silhouette_score we can identify the correct cluster is K=3 since cluster 4 has the max value.

Now we can build our model with K=4

```python
from sklearn.cluster import KMeans
Km_cluster=KMeans(n_clusters=3,random_state=10)
Km_cluster.fit(features_scaled)

```

```python
pred=Km_cluster.fit_predict(features_scaled)
```

```python
cluster=data.copy()
```

```python
cluster['Cluster']=Km_cluster.predict(features_scaled)
#Loading the predicted values in new column
```

```python
cluster
```
